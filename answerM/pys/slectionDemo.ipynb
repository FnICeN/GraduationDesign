```python
import jieba
import numpy as np
from gensim import corpora,models,similarities
import pandas as pd
```

```python
# doc0 = "我不喜欢上海"
# doc1 = "上海是一个好地方"
# doc2 = "北京是一个好地方"
df = pd.read_csv("D:\\GraduationDesign\\语料库\\客服语料\\整理后\\1.csv")
```

```python
# 加载停用词
stop_words = pd.read_csv("D:\\GraduationDesign\\语料库\\stopwords.dat", delimiter="\t", header=None, quoting=3, encoding='utf-8')
stop_words = stop_words[0].tolist()
stop_words
```

```python
# 分词
all_doc = []
# for doc in [doc0, doc1, doc2]:
#     all_doc.append([word for word in jieba.cut(doc)])
# 将df第一列的数据转换为list
for doc in df.iloc[:,0]:
    temp_list = [word for word in jieba.cut(doc) if word not in stop_words and word !=' ']
    all_doc.append(temp_list)
all_doc[0]
```

```python
# 生成词典
dic = corpora.Dictionary(all_doc)
dic.token2id
```

```python
# 生成语料库
corpus = [dic.doc2bow(doc) for doc in all_doc]
corpus
```

```python
tfidf = models.TfidfModel(corpus)
print("corpus：")
for i in tfidf[corpus]:
    print(i)
```

```python
# 计算特征数、相似度索引
featureNum=len(dic.token2id.keys())
featureNum
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=featureNum)
```

```python
# 定义待匹配文档
doc_test = "买的货物有问题怎么办？"
```

```python
# 测试组分词，转词袋
test_doc = [word for word in jieba.cut(doc_test)]
print(test_doc)
test_vec = dic.doc2bow(test_doc)
print(test_vec)
print("测试文本：", tfidf[test_vec])
```

```python
sim = index[tfidf[test_vec]]
```

```python
sim.max()
```

```python
prob_ans = sorted(enumerate(sim), key=lambda item: -item[1])
```

```python
prob_ans
```

```python
all_answer = df.iloc[:,1]
print("问题：",doc_test)
print("匹配到问题：", df.iloc[prob_ans[0][0],0])
print("匹配到问题的关键词：", all_doc[prob_ans[0][0]])
print("回答：", all_answer[prob_ans[0][0]])
```

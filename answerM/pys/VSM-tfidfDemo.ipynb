VSM-tfidf算法通过tfidf对各词向量化并拼接以表示句向量，然后计算两个向量之间的余弦距离等来得出相似度

```python
import jieba
import pandas as pd
import numpy as np
from gensim import corpora, models
path = "D:"
# docs = ["我喜欢吃苹果和梨，还有橙子", "我不喜欢吃香蕉和苹果", "我不喜欢吃橘子和橙子", "我喜欢吃橙子和苹果"]
```

```python
# 读取停用词
stop_words = pd.read_csv(path + "\\GraduationDesign\\语料库\\stopwords.dat", delimiter="\t", header=None, quoting=3, encoding='utf-8')
stop_words = stop_words[0].tolist()
# 读取客服数据
df = pd.read_csv(path + "\\GraduationDesign\\语料库\\客服语料\\整理后\\1.csv")
```

```python
# 分词处理，存入docs
docs = []
for doc in df.iloc[:,0]:
    temp_list = [word for word in jieba.cut(doc) if word not in stop_words and word !=' ']
    docs.append(temp_list)
docs
```

```python
# 生成字典
dictionary = corpora.Dictionary(docs)
# 生成语料库
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
print("原文本：")
for i in docs:
    print(i)
```

```python
print('=================dictinary=============')
print('词ID到这个词在多少篇文档数的映射(dfs):',dictionary.dfs)
print('词到id编码的映射(token2id):',dictionary.token2id)
print('id编码到词的映射(id2token):',dictionary.id2token)
print('处理的文档数(num_docs):',dictionary.num_docs)
print('没有去重词条总数(num_pos):',dictionary.num_pos)
print('对文档内去重后的词条总数，文档间相同词不去重，只要记录BOW矩阵的非零元素个数(num_nnz):',dictionary.num_nnz)
print('=================dictinary=============')
```

```python
print('原词袋表示：')
for c in corpus:
    print(c)
tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
print('转换整个语料库：')
for doc in corpus_tfidf:
    print(doc)
```

```python
test = "如何注册新账户啊？"
test_list = [word for word in jieba.cut(test) if word not in stop_words]
print(test_list)
test_corpus = dictionary.doc2bow(test_list)
print(test_corpus)
```

```python
# 接收待对比的两个corpus
def sim(doc, test):
    vocs = set()
    for i in doc:
        vocs.add(i[0])
    for i in test:
        vocs.add(i[0])
    vocs = list(vocs)
    # corpus计算tfidf
    doc_tfidf = tfidf[doc]
    test_tfidf = tfidf[test]
    # 创建向量
    vec_doc = []
    vec_test = []
    # 向量化
    for i in vocs:
        doc_flag = False
        test_flag = False
        # 词在doc中的tfidf
        for j in doc_tfidf:
            if j[0] == i:
                doc_flag = True
                vec_doc.append(j[1])
                break
        # 词在test中的tfidf
        for j in test_tfidf:
            if j[0] == i:
                test_flag = True
                vec_test.append(j[1])
                break
        # 以0填充
        if not doc_flag:
            vec_doc.append(0)
        if not test_flag:
            vec_test.append(0)
    # 计算余弦相似度
    return np.dot(vec_doc, vec_test) / (np.linalg.norm(vec_doc) * np.linalg.norm(vec_test))
```

```python
# 逐个比较，选出前三最相似者的索引
def compare(test_corpus):
    res = []
    for i in range(len(corpus)):
        res.append([sim(corpus[i], test_corpus), i])
    res.sort(reverse=True)
    return res[:3]
```

```python
prob_ans = compare(test_corpus)
prob_ans
```

```python
all_answer = df.iloc[:,1]
print("问题：", test)
print("匹配到问题：", df.iloc[prob_ans[0][1], 0])
print("回答：", all_answer[prob_ans[0][1]])
```
